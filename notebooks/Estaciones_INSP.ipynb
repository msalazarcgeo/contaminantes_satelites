```python
import geopandas as gpd
import pandas as pd
import shapely
```

```python
rm_quote= lambda x: x.replace('\'', '')
```

```python
df_estaciones=pd.read_csv('../data_s/Estaciones/Coord_estaciones_SINAICA.csv')
df_estaciones=df_estaciones.rename(columns=rm_quote)
for column in ['FUENTE', 'NOM_EST', 'CVE_EST', 'NOM_ENT','NOM_MUN'] :
    df_estaciones[column] = df_estaciones[column].apply(rm_quote)
df_estaciones.head()
```

```python
gdf_estaciones= gpd.GeoDataFrame(df_estaciones)
gdf_estaciones.geometry= gdf_estaciones.apply(lambda x: shapely.geometry.Point(x.LONG, x.LAT,), axis=1)
gdf_estaciones.set_crs(crs='EPSG:4326', inplace=True)
```

```python
gdf_estaciones.crs
```

```python
gdf_estaciones.to_file('../data_s/Estaciones/Coord_estaciones_SINAICA.shp', encoding = 'utf-8')
```

```python
gdf_estaciones_AMM = gdf_estaciones[gdf_estaciones['CVE_EST'].apply(lambda l: 'AMM' in l )]
gdf_estaciones_ZMVM= gdf_estaciones[gdf_estaciones['CVE_EST'].apply(lambda l: 'ZMVM' in l )]
gdf_estaciones_ZMG= gdf_estaciones[gdf_estaciones['CVE_EST'].apply(lambda l: 'ZMG' in l )]
gdf_estaciones_ZMVT= gdf_estaciones[gdf_estaciones['CVE_EST'].apply(lambda l: 'ZMVT' in l )]
```

```python
gdf_estaciones_AMM.to_file('../data_s/Estaciones/Coord_estaciones_SINAICA_AMM.shp', encoding = 'utf-8') 
gdf_estaciones_ZMVM.to_file('../data_s/Estaciones/Coord_estaciones_SINAICA_ZMVM.shp', encoding = 'utf-8')
gdf_estaciones_ZMG.to_file('../data_s/Estaciones/Coord_estaciones_SINAICA_ZMG.shp', encoding = 'utf-8')
gdf_estaciones_ZMVT.to_file('../data_s/Estaciones/Coord_estaciones_SINAICA_ZMVT.shp', encoding = 'utf-8')
```
