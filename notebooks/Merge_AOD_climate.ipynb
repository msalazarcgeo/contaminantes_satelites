## Vamos a unir todo 

```python
def unir_data_AOD_Climate(aod_df, climate_df):
    """Se unen los datos de aodanuales con 
    los climaticos anuales por zona metropolitana"""
    cols_use = climate_ZMVM.columns.difference(aod_df.columns).tolist()
    cols_use= cols_use + ['CVE_EST', 'Mexico_city_hour',
                            'Mexico_city_date', 'timestamp']
    aod_climate  = aod_df.merge( climate_df[cols_use] ,
                    how = 'inner' ,
                    left_on=['CVE_EST', 'Mexico_city_hour',	'Mexico_city_date'],
                    right_on=['CVE_EST', 'Mexico_city_hour',	'Mexico_city_date'],
                    suffixes=('_aod', '_climate')
                    )
    aod_climate.dropna(subset='temperature_2m')
    aod_climate.reset_index(inplace=True, drop=True)
    return aod_climate
    
```

```python
def load_climate(path):
    """Carga y pone las cosas en orden para unir con AOD"""
    df_climate= pd.read_csv(path)
    df_climate['timestamp'] = pd.to_datetime(df_climate['timestamp'])
    df_climate['Mexico_city_time']= df_climate['timestamp'].dt.tz_localize('UTC').dt.tz_convert('America/Mexico_City')
    
    df_climate['Mexico_city_hour'] = df_climate['Mexico_city_time'].apply(lambda l: l.hour)
    df_climate['Mexico_city_date'] = df_climate['Mexico_city_time'].apply(lambda l: l.date)
    
    return df_climate

def load_aod(path):
    """Carga y pone las cosas en orden para unir con climate"""
    df_aod= pd.read_csv(path)
    df_aod= df_aod.dropna(subset='Optical_Depth_047', axis=0)
    df_aod['timestamp'] = pd.to_datetime(df_aod['timestamp'])
    df_aod['Mexico_city_time']= df_aod['timestamp'].dt.tz_localize('UTC').dt.tz_convert('America/Mexico_City')
    
    df_aod['Mexico_city_hour'] = df_aod['Mexico_city_time'].apply(lambda l: l.hour)
    df_aod['Mexico_city_date'] = df_aod['Mexico_city_time'].apply(lambda l: l.date)
    df_aod['AOD']= df_aod['Optical_Depth_047']
    return df_aod

    
    
```

```python
path_aod= '../data_s/AOD_ZM_Mexico/AOD_estacione_met_Mexico_2015.csv'
pat_climate = '../data_s/Climate/ZMVM/ZMVM_est_ERA5_LAND_HOURLY_2015.csv'
df_aod_climate = unir_data_AOD_Climate(load_aod(path_aod) ,
                                       load_climate(pat_climate))
```

```python
df_aod_climate.to_csv('../outputs/ZMVM_2015.csv')
```
